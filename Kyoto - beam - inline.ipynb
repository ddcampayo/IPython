# The beam

Slight redefinition (technical issues):

```python
%pylab inline
```

```python
N=100 ; d = 1.0 / ( N + 3)
```

```python
D4 =( diagflat( -6 * ones(N) ) +
      diagflat(  4 * ones(N-1) ,  1 ) +
      diagflat(  4 * ones(N-1) , -1 ) +
      diagflat( -1 * ones(N-2) ,  2 ) +
      diagflat( -1 * ones(N-2) , -2 )
      )/d**4
```

Also technical: this is a double clamped beam

### Uniform loading

```python
load= ones(N)
```

```python
y=linalg.solve(D4,load)
```

```python
x=linspace(2*d,1-2*d,N)
```

```python
clf(); plot(x,y,'o',x,- x**2 * (1-x)**2 /24)
```

### Complicated loading

```python
def complexload( N ) :
    "Some complicated load vector"
    ll = empty( N) ;
    d=1.0/N
    x=linspace(d,1-d,N)
    ll=exp(- ( (x-0.7) / 0.1) ** 2)
    return ll
```

```python
load2=complexload( N )
```

```python
clf(); plot(x,load2)
```

```python
y=linalg.solve(D4,load2)
```

```python
clf(); plot(x,y)
```

### Central loading

```python
load3=zeros(N); load3[N/2]= 1.0/d
```

```python
y=linalg.solve(D4,load3)
```

```python
clf(); plot(x,y,'o',x, 1.0/12.0 *( -fabs( x-0.5 )*(x-0.5)**2 + (3.0/4.0)*(x-0.5)**2 -1.0/16.0))
```
